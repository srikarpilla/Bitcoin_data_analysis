```python
fg = pd.read_csv("/content/fear_greed_index.csv")
trades = pd.read_csv("/content/historical_data.csv")
```

```python

fg["date"] = pd.to_datetime(fg["date"]).dt.date
fg = fg[["date", "classification", "value"]]

trades["date"] = pd.to_datetime(trades["Timestamp IST"], format="%d-%m-%Y %H:%M").dt.date
```

```python

# MERGE TRADES WITH SENTIMENT ON DATE

data = trades.merge(fg, on="date", how="left")
print("Trades with no sentiment match:", data["classification"].isna().sum())

data = data.dropna(subset=["classification"])

data["is_win"] = data["Closed PnL"] > 0
data["is_loss"] = data["Closed PnL"] < 0
data["is_closed_trade"] = data["Closed PnL"] != 0

order = ["Extreme Fear", "Fear", "Neutral", "Greed", "Extreme Greed"]


#  OVERALL SENTIMENT SPREAD IN THE DATA

print("\n--- Sentiment day counts in trade data ---")
print(data.groupby("classification")["date"].nunique().reindex(order))
```

```python


#  PNL AND WIN RATE BY SENTIMENT

closed_only = data[data["is_closed_trade"]]

pnl_summary = closed_only.groupby("classification")["Closed PnL"].agg(["count", "sum", "mean", "median"])
pnl_summary = pnl_summary.reindex(order)
pnl_summary["win_rate_%"] = closed_only.groupby("classification")["is_win"].mean().reindex(order) * 100

print("\n--- PnL & Win Rate by Sentiment ---")
print(pnl_summary)
```

```python

# TRADING BEHAVIOUR BY SENTIMENT (leverage proxy, size, side)

data["is_long"] = data["Side"] == "BUY"

behaviour = data.groupby("classification").agg(
    avg_size_usd=("Size USD", "mean"),
    avg_fee=("Fee", "mean"),
    long_pct=("is_long", "mean"),
    total_volume_usd=("Size USD", "sum"),
    trade_count=("Size USD", "count"),
).reindex(order)
behaviour["long_pct"] = behaviour["long_pct"] * 100

print("\n--- Trading Behaviour by Sentiment ---")
print(behaviour)

print("\nDone. Summary + merged files saved to outputs/")
```

```python

#  STATISTICAL TEST - IS FEAR PNL DIFFERENT FROM GREED PNL?

fear_pnl = closed_only[closed_only["classification"].isin(["Fear", "Extreme Fear"])]["Closed PnL"]
greed_pnl = closed_only[closed_only["classification"].isin(["Greed", "Extreme Greed"])]["Closed PnL"]

u_stat, p_value = stats.mannwhitneyu(fear_pnl, greed_pnl, alternative="two-sided")
print(f"\nMann-Whitney U test between Fear and Greed PnL: U={u_stat:.0f}, p={p_value:.5f}")
if p_value < 0.05:
    print("Result: statistically significant difference between Fear and Greed PnL")
else:
    print("Result: no statistically significant difference between Fear and Greed PnL")

```

```python

#  ACCOUNT LEVEL VIEW - WHO PROFITS IN WHICH REGIME

account_pnl = closed_only.groupby("Account")["Closed PnL"].sum().sort_values(ascending=False)
print("\n--- Total PnL by Account (top 7) ---")
print(account_pnl)

account_regime = closed_only.groupby(["Account", "classification"])["Closed PnL"].sum().unstack().reindex(columns=order)
print("\n--- PnL by Account and Sentiment ---")
print(account_regime)

# win rate per account in fear vs greed
account_wr = closed_only.groupby(["Account", "classification"])["is_win"].mean().unstack().reindex(columns=order)
print("\n--- Win Rate by Account and Sentiment ---")
print(account_wr)


# COIN LEVEL VIEW

coin_pnl = closed_only.groupby(["Coin", "classification"])["Closed PnL"].sum().unstack().reindex(columns=order)
top_coins = closed_only.groupby("Coin")["Size USD"].sum().sort_values(ascending=False).head(10).index
print("\n--- PnL by Top 10 Coins across Sentiment ---")
print(coin_pnl.loc[coin_pnl.index.intersection(top_coins)])

```

```python

# CHARTS

plt.figure(figsize=(9, 5))
sns.boxplot(data=closed_only, x="classification", y="Closed PnL", order=order, showfliers=False)
plt.axhline(0, color="gray", linestyle="--")
plt.title("Closed PnL Spread across Sentiment Regimes")
plt.tight_layout()
plt.savefig("outputs/pnl_boxplot.png", dpi=150)
plt.close()

plt.figure(figsize=(9, 5))
pnl_summary["win_rate_%"].plot(kind="bar", color="teal")
plt.title("Win Rate % by Sentiment")
plt.ylabel("Win Rate %")
plt.tight_layout()
plt.savefig("outputs/winrate_bar.png", dpi=150)
plt.close()

plt.figure(figsize=(9, 5))
behaviour["long_pct"].plot(kind="bar", color="green")
plt.axhline(50, color="gray", linestyle="--")
plt.title("Long Trade % by Sentiment (above 50 = long-biased)")
plt.ylabel("% Long Trades")
plt.tight_layout()
plt.savefig("outputs/long_bias_bar.png", dpi=150)
plt.close()

plt.figure(figsize=(9, 5))
behaviour["avg_size_usd"].plot(kind="bar", color="purple")
plt.title("Average Trade Size (USD) by Sentiment")
plt.ylabel("Avg Size USD")
plt.tight_layout()
plt.savefig("outputs/avg_size_bar.png", dpi=150)
plt.close()

daily_pnl = closed_only.groupby("date")["Closed PnL"].sum().reset_index()
daily_pnl["date"] = pd.to_datetime(daily_pnl["date"])
daily_pnl = daily_pnl.sort_values("date")
daily_pnl["cum_pnl"] = daily_pnl["Closed PnL"].cumsum()

plt.figure(figsize=(11, 5))
plt.plot(daily_pnl["date"], daily_pnl["cum_pnl"], color="black")
plt.title("Cumulative Closed PnL Over Time (all accounts)")
plt.ylabel("Cumulative PnL (USD)")
plt.tight_layout()
plt.savefig("outputs/cumulative_pnl.png", dpi=150)
plt.close()

plt.figure(figsize=(9, 5))
sns.heatmap(account_wr * 100, annot=True, fmt=".0f", cmap="RdYlGn", center=50)
plt.title("Win Rate % by Account and Sentiment")
plt.tight_layout()
plt.savefig("outputs/account_sentiment_heatmap.png", dpi=150)
plt.close()

print("\nAll charts saved inside outputs/ folder")

```
